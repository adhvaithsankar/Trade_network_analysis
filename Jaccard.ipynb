```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt


# Load dataset
data = pd.read_csv("edge_data_2018.csv")


# Create a bipartite graph: One set for countries, another set for sections
B = nx.Graph()

# Add nodes from both sets
countries = set(data["exporter_id"]).union(set(data["importer_id"]))
sections = set(data["section_no"])

B.add_nodes_from(countries, bipartite=0)  # Countries
B.add_nodes_from(sections, bipartite=1)  # Sections

# Add edges with trade value as weight
for _, row in data.iterrows():
    B.add_edge(row["exporter_id"], row["section_no"], weight=row["value"])
    B.add_edge(row["importer_id"], row["section_no"], weight=row["value"])

# Visualize the bipartite graph
plt.figure(figsize=(12, 10))
pos = nx.spring_layout(B, seed=42)
nx.draw(B, pos, with_labels=True, node_size=300, node_color="skyblue", font_size=10, width=2, edge_color='gray')
plt.title("Bipartite Graph: Countries and Sections", fontsize=16)
plt.show()

# Project the bipartite graph onto the country nodes (monopartite graph)
country_graph = nx.bipartite.projected_graph(B, countries)

# Visualize the country-country graph
plt.figure(figsize=(12, 10))
pos = nx.spring_layout(country_graph, seed=42)
nx.draw(country_graph, pos, with_labels=True, node_size=300, node_color="lightgreen", font_size=10, width=2, edge_color='gray')
plt.title("Monopartite Graph: Countries Interacting Through Sections", fontsize=16)
plt.show()

# Apply network analysis (e.g., centrality)
centrality = nx.degree_centrality(country_graph)
print("\nDegree Centrality of Country-Country Graph:")
print(centrality)
```

```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
from sklearn.metrics import average_precision_score, precision_recall_curve, roc_curve, auc
from sklearn.model_selection import train_test_split

# Load dataset
data = pd.read_csv("edge_data_2018.csv")

# Create a bipartite graph: One set for countries, another set for sections
B = nx.Graph()

# Add nodes from both sets
countries = set(data["exporter_id"]).union(set(data["importer_id"]))
sections = set(data["section_no"])

B.add_nodes_from(countries, bipartite=0)  # Countries
B.add_nodes_from(sections, bipartite=1)  # Sections

# Add edges with trade value as weight
for _, row in data.iterrows():
    B.add_edge(row["exporter_id"], row["section_no"], weight=row["value"])
    B.add_edge(row["importer_id"], row["section_no"], weight=row["value"])

# Project the bipartite graph onto the country nodes (monopartite graph)
country_graph = nx.bipartite.projected_graph(B, countries)

# Split the graph into train and test sets by removing edges for testing
edges = list(country_graph.edges())
train_edges, test_edges = train_test_split(edges, test_size=0.2, random_state=42)

# Remove edges in the test set from the training graph
train_graph = country_graph.copy()
train_graph.remove_edges_from(test_edges)

# Link Prediction: Common Neighbors
def common_neighbors_prediction(graph, node1, node2):
    # Common neighbors score between two nodes
    common_neighbors = list(nx.common_neighbors(graph, node1, node2))
    return len(common_neighbors)

# Link Prediction: Jaccard Similarity
def jaccard_similarity(graph, node1, node2):
    # Jaccard similarity score between two nodes
    return next(nx.jaccard_coefficient(graph, [(node1, node2)]), (0, 0, 0))[2]  # Get Jaccard score

# Get the similarity scores for the test set
common_neighbors_scores = []
jaccard_scores = []
actual_edges = []

for node1, node2 in test_edges:
    actual_edges.append(1)  # This is a true edge
    common_neighbors_scores.append(common_neighbors_prediction(train_graph, node1, node2))
    jaccard_scores.append(jaccard_similarity(train_graph, node1, node2))

# Generate random negative samples for non-edges
non_edges = list(nx.non_edges(train_graph))
for node1, node2 in non_edges[:len(test_edges)]:
    actual_edges.append(0)  # This is a non-edge
    common_neighbors_scores.append(common_neighbors_prediction(train_graph, node1, node2))
    jaccard_scores.append(jaccard_similarity(train_graph, node1, node2))

# Evaluate using Average Precision or Precision-Recall Curve
def plot_roc_pr_curve(true_labels, scores, method_name):
    fpr, tpr, _ = roc_curve(true_labels, scores)
    roc_auc = auc(fpr, tpr)

    precision, recall, _ = precision_recall_curve(true_labels, scores)
    pr_auc = auc(recall, precision)

    plt.figure(figsize=(10, 6))
    plt.plot(fpr, tpr, color='blue', label=f'ROC curve (area = {roc_auc:.2f})')
    plt.plot(recall, precision, color='green', label=f'PR curve (area = {pr_auc:.2f})')
    plt.xlabel('False Positive Rate (FPR)')
    plt.ylabel('True Positive Rate (TPR) / Recall')
    plt.title(f'ROC and PR Curve for {method_name}')
    plt.legend(loc='lower right')
    plt.show()

    return roc_auc, pr_auc

# Plot ROC and Precision-Recall curves
roc_auc_common_neighbors, pr_auc_common_neighbors = plot_roc_pr_curve(actual_edges, common_neighbors_scores, 'Common Neighbors')
roc_auc_jaccard, pr_auc_jaccard = plot_roc_pr_curve(actual_edges, jaccard_scores, 'Jaccard Similarity')

print(f"Common Neighbors - ROC AUC: {roc_auc_common_neighbors}, PR AUC: {pr_auc_common_neighbors}")
print(f"Jaccard Similarity - ROC AUC: {roc_auc_jaccard}, PR AUC: {pr_auc_jaccard}")
```

```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import average_precision_score, precision_recall_curve, roc_curve, auc
from sklearn.model_selection import train_test_split

# Load dataset
data = pd.read_csv("edge_data_2018.csv")

# Create a simplified graph (only one edge between any two countries)
simplified_graph = nx.Graph()

# Add edges with trade value as weight
for _, row in data.iterrows():
    simplified_graph.add_edge(row["exporter_id"], row["importer_id"], weight=row["value"])

# Check if the graph was simplified (should only have one edge between two countries)
print(f"Number of nodes: {simplified_graph.number_of_nodes()}")
print(f"Number of edges: {simplified_graph.number_of_edges()}")

# Split the graph into train and test sets by removing edges for testing
edges = list(simplified_graph.edges())
train_edges, test_edges = train_test_split(edges, test_size=0.2, random_state=42)

# Remove edges in the test set from the training graph
train_graph = simplified_graph.copy()
train_graph.remove_edges_from(test_edges)

# Link Prediction Methods (Common Neighbors, Jaccard Similarity)

def common_neighbors_prediction(graph, node1, node2):
    """Compute the number of common neighbors between two nodes."""
    common_neighbors = list(nx.common_neighbors(graph, node1, node2))
    return len(common_neighbors)

def jaccard_similarity(graph, node1, node2):
    """Compute the Jaccard similarity between two nodes."""
    return next(nx.jaccard_coefficient(graph, [(node1, node2)]), (0, 0, 0))[2]

# Get similarity scores for the test set
common_neighbors_scores = []
jaccard_scores = []

actual_edges = []

# Compute scores for edges in the test set
for node1, node2 in test_edges:
    actual_edges.append(1)  # True edge
    common_neighbors_scores.append(common_neighbors_prediction(train_graph, node1, node2))
    jaccard_scores.append(jaccard_similarity(train_graph, node1, node2))

# Generate random negative samples for non-edges
non_edges = list(nx.non_edges(train_graph))
for node1, node2 in non_edges[:len(test_edges)]:
    actual_edges.append(0)  # Non-edge
    common_neighbors_scores.append(common_neighbors_prediction(train_graph, node1, node2))
    jaccard_scores.append(jaccard_similarity(train_graph, node1, node2))

# Define function to plot evaluation metrics
def plot_roc_pr_curve(true_labels, scores, method_name):
    """Plot ROC and Precision-Recall curves and return AUC scores."""
    fpr, tpr, _ = roc_curve(true_labels, scores)
    roc_auc = auc(fpr, tpr)

    precision, recall, _ = precision_recall_curve(true_labels, scores)
    pr_auc = auc(recall, precision)

    plt.figure(figsize=(10, 6))
    plt.plot(fpr, tpr, color='blue', label=f'ROC curve (AUC = {roc_auc:.2f})')
    plt.plot(recall, precision, color='green', label=f'PR curve (AUC = {pr_auc:.2f})')
    plt.xlabel('False Positive Rate (FPR)')
    plt.ylabel('True Positive Rate (TPR) / Recall')
    plt.title(f'ROC and PR Curve for {method_name}')
    plt.legend(loc='lower right')
    plt.show()

    return roc_auc, pr_auc

# Plot and evaluate all metrics
roc_auc_common_neighbors, pr_auc_common_neighbors = plot_roc_pr_curve(actual_edges, common_neighbors_scores, 'Common Neighbors')
roc_auc_jaccard, pr_auc_jaccard = plot_roc_pr_curve(actual_edges, jaccard_scores, 'Jaccard Similarity')

# Create a DataFrame for easy heatmap plotting
metrics_df = pd.DataFrame({
    'Method': ['Common Neighbors', 'Jaccard Similarity'],
    'ROC AUC': [roc_auc_common_neighbors, roc_auc_jaccard],
    'PR AUC': [pr_auc_common_neighbors, pr_auc_jaccard]
})

# Set Method column as index
metrics_df.set_index('Method', inplace=True)

# Plot heatmap using Seaborn
plt.figure(figsize=(8, 6))
sns.heatmap(metrics_df.T, annot=True, cmap='coolwarm', fmt=".2f", cbar=True, center=0)
plt.title("ROC AUC and PR AUC Heatmap")
plt.show()

# Print results for all methods
print(f"Common Neighbors - ROC AUC: {roc_auc_common_neighbors:.2f}, PR AUC: {pr_auc_common_neighbors:.2f}")
print(f"Jaccard Similarity - ROC AUC: {roc_auc_jaccard:.2f}, PR AUC: {pr_auc_jaccard:.2f}")
```
